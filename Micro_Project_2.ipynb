```python
# --- MICRO-PROJECT 2: Data Visualization (Report Step) ---

# Ensure visualizations appear inline
%matplotlib inline

# Import libraries if not already imported
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset if not already loaded
df = pd.read_csv("Employee Attrition.csv")

# Drop irrelevant or constant columns
columns_to_drop = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']
df = df.drop(columns=columns_to_drop)

# Set a visual theme
sns.set(style="whitegrid")

# 1. Attrition Count Plot
sns.countplot(data=df, x="Attrition", palette="Set2")
plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.show()

# 2. Attrition vs DistanceFromHome (Boxplot)
sns.boxplot(data=df, x="Attrition", y="DistanceFromHome", palette="Set3")
plt.title("Attrition vs Distance From Home")
plt.xlabel("Attrition")
plt.ylabel("Distance From Home")
plt.show()

# 3. Job Satisfaction vs Attrition (Violin Plot)
sns.violinplot(data=df, x="Attrition", y="JobSatisfaction", palette="Set1")
plt.title("Attrition vs Job Satisfaction")
plt.xlabel("Attrition")
plt.ylabel("Job Satisfaction")
plt.show()

# 4. Years at Company vs Attrition (Histogram)
sns.histplot(data=df, x="YearsAtCompany", hue="Attrition", multiple="stack", palette="pastel", bins=15)
plt.title("Years at Company by Attrition Status")
plt.xlabel("Years at Company")
plt.ylabel("Number of Employees")
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(12, 10))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
